```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
train_data = pd.read_csv('/content/get_out_cleaned (1).csv')
train_data.head(5)
```

# New section

```python
pd.set_option('max_colwidth',1800)
train_data[train_data.Rating == np.positive][0:5]
train_data[train_data.Rate <= 10][0:5]
```

```python
train_data.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sn
%matplotlib inline
plt.figure(figsize=(10,10))
plt.title("Number of records of positive and negative sentiments")
plot = sn.countplot(x = 'Rate',data=train_data)
for p in plot.patches:
    plot.annotate(p.get_height(),(p.get_x()+0.1 ,p.get_height()+5))
```
